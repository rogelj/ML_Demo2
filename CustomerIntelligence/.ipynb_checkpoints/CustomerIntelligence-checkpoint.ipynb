````python
from sherlockml import customerintelligence
import pandas as pd
df = pd.read_csv('client_postcodes.csv', header=None)
df['random']=1
result = customerintelligence.analyse(df[0], num_jobs=2)
````

Customer Intelligence takes in your customer postcodes and
outputs areas that are demographically similar to where your customers are based.
Under the hood, it uses a novel algorithm developed by ASI Data Science running
on UK census data.

In the map below, yellow areas are demographically more similar to your customer base.
Blue areas are less similar.

SherlockML comes with Python libraries for Customer Intelligence.
To get started,

```python
from sherlockml import customerintelligence
```

````python
import matplotlib as mpl
import matplotlib.pyplot as plt
def show_colormap(cmap, show_axis=False):
    fig = plt.figure(figsize=(8, 3))
    ax = fig.add_axes([0.05, 0.80, 0.9, 0.15])
    cb = mpl.colorbar.ColorbarBase(ax, cmap=cmap,
                                   orientation='horizontal')
    if not show_axis:
        ax.get_xaxis().set_ticks([])
    plt.show()
show_colormap(mpl.cm.viridis)
````
